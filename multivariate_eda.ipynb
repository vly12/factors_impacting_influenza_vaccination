```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

pd.options.display.float_format = '{:,.4f}'.format
```

```python

cohorts = [
    "WEAKENED_IMMUNE_SYSTEM",
    "AGE_50_TO_64",
    "AGE_65_TO_74",
    "AGE_75_AND_OVER",
    "ALL_HEALTH_CARE_WORKERS",
    "ALL_SOCIAL_CARE_WORKERS",
    "18_TO_64_FLU_AT_RISK",
    "OLDER_PEOPLE_CARE_HOME"
]

df_sample = pd.read_parquet("INPUT DATA PATH HERE")
```

```python

for cohort in cohorts:
    cohort_df = df_sample[df_sample["cohort_group_ML_analysis"] == cohort]
    print(f"Summary statistics for {cohort}:\n", cohort_df.describe())
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    fig.suptitle(f'Distribution Plots for {cohort}', fontsize=14)

    #Age distribution
    sns.histplot(cohort_df['patient_age'], bins=20, ax=axes[0, 0])
    axes[0, 0].set_title('Age distribution')
    age_skew = skew(cohort_df['patient_age'])
    axes[0, 0].text(0.95, 0.95, f"Skew: {age_skew:.2f}", transform=axes[0, 0].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.6))

    #Sex distribution
    sns.countplot(x='patient_sex', data=cohort_df, ax=axes[0, 1])
    axes[0, 1].set_title('Sex distribution')
    axes[0, 1].set_xticklabels(['Male', 'Female'])
    sex_skew = skew(cohort_df['patient_sex'])
    axes[0, 1].text(0.95, 0.95, f"Skew: {sex_skew:.2f}", transform=axes[0, 1].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.6))

    #SIMD quintile
    sns.countplot(x='SIMD_quintile', data=cohort_df, ax=axes[0, 2])
    axes[0, 2].set_title('SIMD quintile')
    simd_skew = skew(cohort_df['SIMD_quintile'])
    axes[0, 2].text(0.95, 0.95, f"Skew: {simd_skew:.2f}", transform=axes[0, 2].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.6))

    #UR8
    sns.countplot(x='UR8_2022', data=cohort_df, ax=axes[1, 0])
    axes[1, 0].set_title('Urban-Rural 8')
    ur8_skew = skew(cohort_df['UR8_2022'])
    axes[1, 0].text(0.95, 0.95, f"Skew: {ur8_skew:.2f}", transform=axes[1, 0].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.6))

    #Vaccination attendance
    sns.countplot(x='attended_vaccination_event', data=cohort_df, ax=axes[1, 1])
    axes[1, 1].set_title('Attended Vaccination Distribution')
    axes[1, 1].set_xticklabels(['Did Not Attend', 'Attended'])
    attend_skew = skew(cohort_df['attended_vaccination_event'])
    axes[1, 1].text(0.95, 0.95, f"Skew: {attend_skew:.2f}", transform=axes[1, 1].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.6))

    #Hide unused subplot
    axes[1, 2].axis('off')

    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig(f"{cohort}_distribution_plots.png")

    plt.show()
```

```python

for cohort in cohorts:
    cohort_df = df_sample[df_sample["cohort_group_ML_analysis"] == cohort].copy()

    # Cramer's V matrix
    def cramers_v(confusion_matrix):
        chi2 = chi2_contingency(confusion_matrix)[0]
        n = confusion_matrix.sum().sum()
        r, k = confusion_matrix.shape
        return np.sqrt(chi2 / (n * (min(k - 1, r - 1))))

    cats = ['patient_sex','SIMD_quintile','UR8_2022','attended_vaccination_event']
    cramer = pd.DataFrame(index=cats, columns=cats)
    for col1 in cats:
        for col2 in cats:
            confusion = pd.crosstab(cohort_df[col1], cohort_df[col2])
            cramer.loc[col1, col2] = cramers_v(confusion)
    cramer = cramer.astype(float)

    # Stacked bar for SIMD vs target
    simd_ct = pd.crosstab(cohort_df['SIMD_quintile'], cohort_df['attended_vaccination_event'], normalize='index')

    # Stacked bar for patient_sex vs target
    patient_sex_ct = pd.crosstab(cohort_df['patient_sex'], cohort_df['attended_vaccination_event'], normalize='index')
    patient_sex_ct.index = patient_sex_ct.index.map({0: 'Male', 1: 'Female'})

    # Line plot for attendance probability by age
    min_age = cohort_df['patient_age'].min()
    max_age = cohort_df['patient_age'].max()
    bins = np.linspace(min_age, max_age, num=9)
    cohort_df['age_bin'] = pd.cut(cohort_df['patient_age'], bins=bins, right=False)
    rate = cohort_df.groupby('age_bin')['attended_vaccination_event'].mean().reset_index()
    rate['age_mid'] = rate['age_bin'].apply(lambda x: x.left + (x.right - x.left) / 2)

    # Create 3x2 grid of plots
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle(f'Multivariate EDA for {cohort}', fontsize=16)

    sns.heatmap(cramer, annot=True, cmap='Greens', fmt='.2f', ax=axes[0, 0])
    axes[0, 0].set_title("Cramer's V Correlational Matrix")

    patient_sex_ct.plot(kind='bar', stacked=True, color=['tomato','steelblue'], ax=axes[0, 1])
    axes[0, 1].legend(['Did not attend','Attended'], title='Outcome')
    axes[0, 1].set_ylabel('Proportion within Sex')
    axes[0, 1].set_title('Attendance proportion by Patient Sex')

    simd_ct.plot(kind='bar', stacked=True, color=['tomato','steelblue'], ax=axes[1, 0])
    axes[1, 0].legend(['Did not attend','Attended'], title='Outcome')
    axes[1, 0].set_ylabel('Proportion within SIMD quintile')
    axes[1, 0].set_title('Attendance proportion by SIMD quintile')

    axes[1, 1].plot(rate['age_mid'], rate['attended_vaccination_event'], marker='o')
    axes[1, 1].set_xlabel('Age')
    axes[1, 1].set_ylabel('P(Attend)')
    axes[1, 1].set_title('Attendance probability by age based on outcome')
    axes[1, 1].set_ylim(0, 1)


    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig(f"{cohort}_multivariate_plots.png")
    plt.show()

    cohort_df.drop(columns='age_bin', inplace=True)
```
