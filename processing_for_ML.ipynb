```python
import pandas as pd
#One hot encode UR8_2022 
df = pd.read_parquet('/PHI_conf/VaccineUptake/Analysts/Vay/vaccinate_uptake_ML_analysis/vaccine_data/master_data/cohort_df_merged.parquet')

df.info()

df
```

```python


# One-hot encode UR8_2022
ur8_encoded = pd.get_dummies(df['UR8_2022'], prefix='UR8_2022').astype(int)

# Find the index of SIMD_quintile
simd_index = df.columns.get_loc('SIMD_quintile')

# Split the DataFrame into parts
before = df.iloc[:, :simd_index + 1]
after = df.iloc[:, simd_index + 1:].drop(columns=['UR8_2022'])

# Concatenate with one-hot columns in the middle
df_final = pd.concat([before, ur8_encoded, after], axis=1)

df_final
```
